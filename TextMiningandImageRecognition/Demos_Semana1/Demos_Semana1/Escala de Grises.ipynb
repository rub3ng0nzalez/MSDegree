```python
import cv2
import numpy as np
import sys

import matplotlib.pyplot as plt
```

```python
img = cv2.imread('paris.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python

ancho = img.shape[0]
alto = img.shape[1]

gray = np.zeros((ancho, alto, 1))
gray2 = np.zeros((ancho, alto, 1))

for i in range(0, ancho):
    for j in range(0, alto):
        pixel = img[i, j]
        ponderado = (int(0.299*pixel[0]) + int(0.587*pixel[1]) + int(0.114*pixel[2]))
        aritmetico = int(int(pixel[0]) + int(pixel[1]) + int(pixel[2]))/3
        gray[i,j] = ponderado
        gray2[i,j] = aritmetico
        
cv2.imwrite('gris_ponderado.jpg', gray)
cv2.imwrite('gris_aritmetico.jpg', gray2)
```

```python
img = cv2.imread('gris_ponderado.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.imread('gris_aritmetico.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## Escala de grises 3D

```python
gris_rojo = np.zeros((ancho, alto, 1))
gris_verde = np.zeros((ancho, alto, 1))
gris_azul = np.zeros((ancho, alto, 1))

for i in range(0, ancho):
    for j in range(0, alto):
        pixel = img[i, j]
        gris_rojo[i,j] = pixel[0]
        gris_verde[i,j] = pixel[1]
        gris_azul[i,j] = pixel[2]

cv2.imwrite('gris_rojo.jpg', gris_rojo)
cv2.imwrite('gris_verde.jpg', gris_verde)
cv2.imwrite('gris_azul.jpg', gris_azul)
```

```python
rojo = cv2.imread('gris_rojo.jpg', cv2.IMREAD_COLOR)
verde = cv2.imread('gris_verde.jpg', cv2.IMREAD_COLOR)
azul = cv2.imread('gris_azul.jpg', cv2.IMREAD_COLOR)

rojo = cv2.cvtColor(rojo, cv2.COLOR_BGR2RGB)
verde = cv2.cvtColor(verde, cv2.COLOR_BGR2RGB)
azul = cv2.cvtColor(azul, cv2.COLOR_BGR2RGB)
```
