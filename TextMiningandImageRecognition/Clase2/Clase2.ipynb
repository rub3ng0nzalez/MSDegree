```python
import cv2
import numpy as np
import sys

import matplotlib.pyplot as plt
```

```python
img_str='CoatepequeShort.jpg'
img = cv2.imread(img_str)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
alto = img.shape[0]
ancho = img.shape[1]

inversa = np.zeros((alto,ancho,1))

for i in range(0,alto):
    for j in range(0,ancho):
        pixel = gray[i,j]
        newPixel= 255 - pixel
        inversa[i,j]=newPixel

cv2.imwrite('inversa.jpg', inversa)
```

```python
img = cv2.imread('inversa.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

## Ecualizacion

```python
img = cv2.imread('image.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
alto = gray.shape[0]
ancho = gray.shape[1]

ecualizada = np.zeros((alto,ancho,1))

frecuencia = np.zeros((256))

for i in range(0,alto):
    for j in range(0,ancho):
        pixel = gray[i,j]
        frecuencia[pixel]=frecuencia[pixel]+1
# pdf - histograma
densidad = np.true_divide(frecuencia, (ancho * alto))

#cdf
cdf = np.cumsum(densidad, dtype = float)

for i in range(0,alto):
    for j in range(0,ancho):
        pixel=gray[i,j]
        newPixel = 255*cdf[pixel]
        ecualizada[i,j]=newPixel

cv2.imwrite('ecualizada.jpg',ecualizada)
```

```python
img = cv2.imread('ecualizada.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ecualizado = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(ecualizado, cmap="gray")
```

```python
histOriginal = cv2.calcHist([gray], [0], None, [256], [0,256])
plt.plot(histOriginal, color='gray')
```

```python
histEcualizado = cv2.calcHist([ecualizado], [0], None, [256], [0,256])
plt.plot(histEcualizado, color='gray')
```
