```python
import numpy as np
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image =cv2.imread("prueba.jpg", cv2.IMREAD_COLOR)
ancho = image.shape[0]
alto = image.shape[1]

grayScale=np.zeros((ancho, alto, 1))

for i in range(0, ancho):
    for j in range(0, alto):
        pixel = image[i,j]
        ponderado = ((0.114*pixel[0]) + (0.587*pixel[1]) + (0.299*pixel[2]))
        grayScale[i,j] = ponderado

cv2.imwrite("gris_ponderado.jpg",grayScale)
```
