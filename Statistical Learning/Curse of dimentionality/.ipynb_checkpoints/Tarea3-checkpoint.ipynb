## Maldicion de la dimensionalidad

#### Ruben González 20003314

La causa común de estos problemas es que cuando aumenta la dimensionalidad, el volumen del espacio aumenta exponencialmente haciendo que los datos disponibles se vuelven dispersos. Esta dispersión es problemática para cualquier método que requiera significación estadística.

### Demostración visual

#### Una dimensión
Supongamos que tenemos 50 datos registrados entre 0 y 100.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
fig = plt.figure()
ax  = plt.axes()
fig.set_size_inches(12, 1)
x = random.sample(range(0, 100), 50) 
y = [0 for xval in x]
plt.scatter(x, y)
# Lineas de segmentación
for grid_pt in [20, 40, 60, 80]:
    plt.axvline(x=grid_pt, color='#D8D8D8')
ax.set_xlim((0,100))
ax.set_xlabel("Dimension #1", fontsize=14)
ax.set_ylabel("")
plt.yticks([], [])
plt.title("1D")
plt.show()
```

Con 5 segmentos en nuestra primera dimensión, existirá un promedio de 50/5 = 10 puntos por celda. Ahora subamos una dimensión:

#### Dos dimensiones

```python
fig = plt.figure()
ax  = plt.axes()
fig.set_size_inches(8, 8)
# Ahora cada observación consiste en 2 dimensiones o 2 features.
x = random.sample(range(0, 100), 50) 
y = random.sample(range(0, 100), 50) 
plt.scatter(x, y)
# Lineas de segmentación del espacio
for grid_pt in [20, 40, 60, 80]:
    plt.axvline(x=grid_pt, color='#D8D8D8')
    plt.axhline(y=grid_pt, color='#D8D8D8')
ax.set_xlim((0,100))
ax.set_ylim((0,100))
ax.set_xlabel("Dimension #1", fontsize=14)
ax.set_ylabel("Dimension #2", fontsize=14)
plt.title("2D")
plt.show()
```

Segmentando el espacio en 5 intervalos por cada dimensión vemos que ahora tendriamos en promedio 50/(5 * 5) = 2 puntos por celda en promedio. De hecho comenzamos a ver algunos de estos segmentos que no cuentan con ninguna observación.

#### Tres dimensiones

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax  = fig.add_subplot(1,1,1,projection='3d')
fig.set_size_inches(10, 8)
# Ahora cada observación tiene 3 dimensiones
x = random.sample(range(0, 100), 50) 
y = random.sample(range(0, 100), 50) 
z = random.sample(range(0, 100), 50)
ax.scatter(x, y, z)
# Líneas de segmentación del espacio
for grid_pt in [20, 40, 60, 80]:
    plt.axvline(x=grid_pt, color='#D8D8D8')
    plt.axhline(y=grid_pt, color='#D8D8D8')
ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_zlim(0,100)
ax.set_xlabel("Dimension #1", fontsize=14)
ax.set_ylabel("Dimension #2", fontsize=14)
ax.set_zlabel("Dimension #3", fontsize=14)
plt.title("3D")
plt.show()
```

Con 5 intervalos por dimensión, tendríamos 50/(5 * 5 * 5) = 0.4 puntos por celda en promedio.

### Conclusión
Entre más dimensiones agregamos, creamos más espacio posible donde pueden haber observaciones. Este espacio no necesariamente se llena correctamente con la data inicial.

La cantidad de data necesaria para soportar los posibles resultados crece exponencialmente mientras aumentamos dimensionalidad.
