### Implementacion de Hard-clustering K-means 

```python
# Funcion para obtener valores random
def InterpolarRandom(minimo,maximo):
    valorRandom=np.random.randint(1,100)
    Pieza=(maximo-minimo)/100
    Posicion=Pieza*valorRandom
    Salida=float(minimo+Posicion)
    return Salida
```

```python
valor=InterpolarRandom(15, 50)
valor
```

```python
def assignment(data, centroids, colmap):
        for i in centroids.keys():
            # sqrt((x1 - x2)^2 - (y1 - y2)^2)
            data['distance_from_{}'.format(i)] = (
                np.sqrt(
                    (data['Estatura(metros)'] - centroids[i][0]) ** 2
                    + (data['Edad(años)'] - centroids[i][1]) ** 2
                )
            )
        centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
        data['closest'] = data.loc[:, centroid_distance_cols].idxmin(axis=1)
        data['closest'] = data['closest'].map(lambda x: int(x.lstrip('distance_from_')))
        data['color'] = data['closest'].map(lambda x: colmap[x])
        return data
```

```python
#Funcion que implementa K-means sin usar skitLearn
def kmeans(kno, semilla, data):
    # Initialization 
    np.random.seed(semilla)
    k = kno
    centroids = dict()
    for i in range(k):
        centroids[i+1]= [InterpolarRandom(1.4, 2), InterpolarRandom(15, 45)]
        
    colmap = {1: 'red', 2: 'green', 3: 'blue', 4: 'yellow', 5: 'purple'}
    
    # Asignation
    data = assignment(data, centroids, colmap)

    # Update Stage
    old_centroids = copy.deepcopy(centroids)

    def update(k):
        for i in centroids.keys():
            centroids[i][0] = np.mean(data[data['closest'] == i]['Estatura(metros)'])
            centroids[i][1] = np.mean(data[data['closest'] == i]['Edad(años)'])
        return k

    centroids = update(centroids)
    # Repetir hasta que ya no cambie 
    while True:
        closest_centroids = data['closest'].copy(deep=True)
        centroids = update(centroids)
        data = assignment(data, centroids, colmap)
        if closest_centroids.equals(data['closest']):
            break
    
    #Calculo de wcss
    SumaCuadrados = []
    for i in range(k):
        distObj=data[data.closest==i+1]
        cuadrados=distObj['distance_from_'+str(i+1)].pow(2)
        SumaCuadrados.append(cuadrados.sum())
    wcss=np.sum(SumaCuadrados)
    return data,centroids,colmap, wcss
```

```python
#Funcion para graficar el modelo
def GraficarKmeans(data, centroids, colmap):
    #Graficar Resultado
    fig = plt.figure(figsize=(5, 5))
    plt.scatter(data['Estatura(metros)'], data['Edad(años)'], color=data['color'], alpha=0.5, edgecolor='k')
    for i in centroids.keys():
        plt.scatter(*centroids[i], color='black')
    plt.xlim(1.4, 2.1)
    plt.ylim(15, 45)
    plt.title('K-means con K='+str(i))
    plt.xlabel('Estatura(metros)')
    plt.ylabel('Edad(años)')
    plt.show()
```

#### Visualizacion de resultados

```python
Error = []
for i in range(1,6):
    SalidaDf, Centroides, Mapa, Wcss = kmeans(i,145,data)
    Error.append(Wcss)
    GraficarKmeans(SalidaDf, Centroides, Mapa)
```

#### Analisis metodo del codo (elbow method)

```python
plt.plot(range(1, 6), Error)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Analizando los resultados, vemos que a partir del cluster 4 en adelante ya no hay una diferencia significativa en la curva por lo que seleccionamos 4 como el K valido para este caso. El metodo del codo es adecuado pues el grafico de WCSS obtenido es lo suficientemente bueno como para encontrar el K optimo a partir de la grafica obtenida. Si los cambios no fueran tan sustanciales entonces si se dificultaria encontrar K y quizas se deberia utilizar otro metodo.

```python
SalidaDf, Centroides, Mapa, Wcss = kmeans(4,145,data)
GraficarKmeans(SalidaDf, Centroides, Mapa)
```

### Metodo Soft-clustering GMM con skitlearn

```python
data=pd.DataFrame(data, columns =["Estatura(metros)","Edad(años)"])
```

```python
# Usamos K=4 segun lo obtenido en el modelo anterior
from sklearn.mixture import GaussianMixture

#Se entrena el modelo
gmm = GaussianMixture(n_components=4).fit(data)
labels = gmm.predict(data)
plt.scatter(data["Estatura(metros)"],data["Edad(años)"], c=labels, s=40, cmap='viridis')
```

```python
dataTest=pd.DataFrame(dataTest, columns =["Estatura(metros)","Edad(años)"])
```

#### Prediccion en los dataTest y probabilidad:

```python
labels = gmm.predict(dataTest)
probs = gmm.predict_proba(dataTest)
plt.scatter(dataTest["Estatura(metros)"],dataTest["Edad(años)"], c=labels, s=40, cmap='viridis')
print(probs.round(3))
```

La diferencia entre ambos resultados obtenidos es que la persona en el datatest que mide 1.82 con 25 años de edad es clasificada en diferente cluster usando el metodo k-means mientras que con el metodo GMM, este es clasificado en el mismo cluster que los otros con estaturas similares.

## Parte 2
- Usar pestaña uno para hacer el entrenamiento del modelo de detección de anomalías.
- Usar pestaña dos y tres como set de validación(seleccionar epsilon) y pruebas.

```python
# Convertir data en np.array para mejor manipulacion con valores mayores a epsilon
Xa = np.array(data)
```

```python
from scipy.stats import multivariate_normal
def estimateGaussian(dataset):
    mu = np.mean(dataset, axis=0)
    sigma = np.cov(dataset.T)
    return mu, sigma
    
def multivariateGaussian(dataset,mu,sigma):
    p = multivariate_normal(mean=mu, cov=sigma)
    return p.pdf(dataset)
```

```python
mu, sigma=estimateGaussian(Xa)
print(mu, sigma)
```

```python
def graficaGauss(epsilon, Xa,p):
    plt.figure(figsize=(8,6))
    plt.scatter(Xa[:,0],Xa[:,1],marker="x",c=p,cmap='viridis');
    # Circling of anomalies
    outliers = np.nonzero(p<epsilon)[0]
    plt.scatter(Xa[outliers,0],Xa[outliers,1],marker="o",facecolor="none",edgecolor="r",s=70)
    return outliers
```

#### Busqueda de anomalias
En el mismo data train que nos sirvio para obtener media y varianza se realizara un analisis usando un epsilon de 0.05

```python
ptrain = multivariateGaussian(Xa, mu, sigma)
graficaGauss(0.05,Xa=Xa,p=ptrain)
```

Los registros 17,21 y 26 se salen del intervalo de confianza establecido y por eso estan marcados con rojo, ahora se realizara la busqueda con dataTest

```python
Xtest = np.array(dataTest)
ptest = multivariateGaussian(Xtest,mu,sigma)
graficaGauss(0.05,Xa=Xtest,p=ptest)
```

El registro 3 se sale del intervalo de confianza establecido, ahora procedemos a hacer la prueba en el dataset de anomalias:

```python
Xanomalia = np.array(dataAnomalias)
pAnomalia = multivariateGaussian(Xanomalia,mu,sigma)
graficaGauss(0.05,Xa=Xanomalia,p=pAnomalia)
```

Como era de esperarse todos los registros del dataset de anomalias estan fuera del intervalo de confianza.
